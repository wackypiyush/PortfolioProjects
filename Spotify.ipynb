```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("SpotifyFeatures.csv")
data.head()
```

```python
#Checking for missing data

data.isnull().sum()
```

```python
data.info()
```

```python
#Least 10 popular songs

data.sort_values(by=['popularity'],inplace=False,ascending=True).head(10)
```

```python
data.describe().transpose()
```

```python
#Top 10 Songs

most_popular=data.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
# Changing duration in ms to sec

data['duration']=data['duration_ms'].apply(lambda x: round(x/1000))
data.drop('duration_ms',axis=1,inplace=True)
data.head()
```

```python
corr_df=data.drop(['key','mode'],axis=1).corr(method='pearson')

plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=data.sample(int(0.004*len(data)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='energy',y='loudness').set(title="Correlation between Loudness and Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='acousticness',y='popularity',color='r').set(title="Correlation between Popularity and Acousticness")
```

```python
plt.figure(figsize=(15,10))
plt.title("Duration of the songs in different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration',data=data)
plt.xlabel('Duration in seconds')
plt.ylabel('Genres')
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,6))
famous=data.sort_values('popularity',ascending=False).head(10)
sns.barplot(x='popularity',y='genre',data=famous).set(title='Top 5 genres by popularity')
```
